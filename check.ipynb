```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dff1=pd.read_csv("rateS.csv")
dff1.sort_values("Literacy",kind="quicksort")
```

```python
l=dff1["Literacy"]
c=[]
for i in l:
    c.append(float(i.split("%")[0]))
```

```python
top=c[:5]
rest=sum(c[5:])
top.append(rest)
city=dff1["District"].values
cit=[city[0],city[1],city[2],city[3],city[4],"Others"]
```

```python
colors=["red","lightskyblue","yellowgreen","lightyellow","brown","lightcoral"]
plt.figure(figsize=(8,6))
patches, texts=plt.pie(c[:6],colors=colors, shadow=True, startangle=0)
plt.legend(patches, cit,loc="best")
plt.axis("equal")
plt.tight_layout()
plt.savefig("pl.jpg")
plt.show()
```
